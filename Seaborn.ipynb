```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
planets = sns.load_dataset('planets')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

# Scatter Plot

```python
sns.scatterplot(x = 'tip', y = 'total_bill', data=tips, hue='day', size='size')
```

# Histogram

```python
sns.histplot(tips['tip'], kde=True, bins=30)
```

# Dis plot

```python
sns.displot(tips['tip'], kde=True, bins=30)
```

# Bar Plot

```python
sns.barplot(x='sex', y='tip', data=tips)
```

# Box plot

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex')
```

# Strip Plot

```python
sns.stripplot(x='day',y='tip', data=tips, hue='sex')
```

# Joint Plot

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```
